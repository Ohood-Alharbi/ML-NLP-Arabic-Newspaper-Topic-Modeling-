```python
start = time.time()

dftext['tokenized_text'] = dftext['T_Text'].apply(word_tokenize) 


end = time.time()


print(end - start)
```

```python
start = time.time()

a = dftext['tokenized_text'].tolist()
aflat =[item for sublist in a for item in sublist]

end = time.time()
print(end - start)
```

```python
## sentence tokenizer
start = time.time()

dftext['sent'] = dftext['T_Text'].apply(sent_tokenize) 


end = time.time()


print(end - start)
```

```python
start = time.time()

b = dftext['sent'].tolist()
bflat =[item for sublist in b for item in sublist]

end = time.time()
print(end - start)
len(bflat)
```

```python
dfwords =pd.DataFrame(columns=['word'],data=aflat)
```

```python
dfwords.to_csv("words.csv",encoding='utf-8-sig')
```

```python
dfsent = pd.DataFrame(columns=['sentence'],data=bflat)
dfsent.to_csv("sent.csv",encoding='utf-8-sig')
```

```python
from collections import Counter
counts = Counter(aflat)
len(counts)
```

```python
dffreq = pd.DataFrame(columns=['word','frequency'],data=zip(counts.keys(),counts.values()))
```

```python
dffreq.to_csv("words_freq.csv",encoding='utf-8-sig')
```
